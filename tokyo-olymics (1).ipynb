```python
plt.figure(figsize=(30,50))
# Bar chart showing average score for racing games by platform
sns.barplot(x=data['Rank'], y=data["Team/NOC"])
# Add label for horizontal axis
plt.xlabel("")
# Add label for vertical axis
plt.title("Average Score for Racing Games, by Platform")

# Check your answer
```

* **barplot() method. A barplot is basically used to aggregate the categorical data according to some methods and by default it's the mean.**

```python
plt.figure(figsize=(50,6))
# Line chart showing the number of visitors to each museum over time
sns.lineplot(x="NOCCode",y="Gold Medal",data=data)
plt.xlabel("country")
plt.title("country")
```

* **This is the lineplot between "NOCCode" and "Gold Medal"**
